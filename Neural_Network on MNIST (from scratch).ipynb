```python
import numpy as np
from sklearn import datasets
```

```python
bc = datasets.load_digits()
X=bc.data
Y=bc.target
X.shape,Y.shape
```

```python
def sig(z):
    return 1/(1 + np.exp(-z))
```

```python
def dsig(z):
    return sig(z)*(1 - sig(z))
```

```python
wh = 2 * np.random.random((64,64)) - 1
wh2 = 2 * np.random.random((64,64)) - 1
bh = np.random.random((1, 64))* 2 - 1
bh2 = np.random.random((1, 64))* 2 - 1
wo = 2 * np.random.random((64,1)) - 1
bo = np.random.random(1)* 2 - 1
```

```python
def forward(x, wh,wh2, wo, bh,bh2, bo):
    output0 = x
    hidden_layer_input = np.dot(output0, wh) + bh
    hidden_layer_output = sig(hidden_layer_input)
    
    output_layer_input = np.dot(hidden_layer_output, wh2) + bh2
    output = sig(output_layer_input)
    
    output_layer_input2 = np.dot(output, wo) + bo
    output2=sig(output_layer_input2)
    return output2
```

```python
def train(X, Y, wh,wh2, wo, bh,bh2, bo, iterations, lr):
    for i in range(iterations):
        output0 = X
        hidden_layer_input = np.dot(output0, wh) + bh
        hidden_layer_output = sig(hidden_layer_input)
        
        hidden_layer_input2 = np.dot(hidden_layer_output, wh2) + bh2
        hidden_layer_output2 = sig(hidden_layer_input2)

        output_layer_input = np.dot(hidden_layer_output2, wo) + bo
        output = sig(output_layer_input)

        first_term_output = output - Y
        second_term_output = dsig(output_layer_input)
        first_two_terms_output = first_term_output * second_term_output
        
        second_term_hidden_layer2 = dsig(hidden_layer_input2)
        first_term_hidden_layer2 = np.dot(first_two_terms_output, wo.T)
        first_two_terms_hidden2 = first_term_hidden_layer2*second_term_hidden_layer2

        second_term_hidden_layer1 = dsig(hidden_layer_input)
        first_term_hidden_layer1 = np.dot(first_two_terms_hidden2, wh2.T)
        first_two_terms_hidden1 = first_term_hidden_layer1*second_term_hidden_layer1

        changes_wo = np.dot(hidden_layer_output2.T, first_two_terms_output)
        changes_wh2 = np.dot(hidden_layer_output.T, first_two_terms_hidden2)
        changes_wh = np.dot(output0.T, first_two_terms_hidden1)

        wh = wh - lr * changes_wh
        wh2 = wh2 - lr* changes_wh2
        wo = wo - lr * changes_wo

        changes_bh = np.sum(first_two_terms_hidden1, axis = 0)
        changes_bh2 = np.sum(first_two_terms_hidden2, axis = 0)
        changes_bo = np.sum(first_two_terms_output, axis = 0)

        bh = bh - lr * changes_bh
        bh2 = bh2 - lr* changes_bh2
        bo = bo - lr*changes_bo
    return wh, wh2, wo, bh, bh2, bo
        
```

```python
wh, wh2, wo, bh, bh2, bo = train(X, Y, wh, wh2, wo, bh, bh2, bo, 1000, 0.01)
```

```python
forward(X, wh,wh2, wo, bh,bh2, bo)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_true = Y,y_pred = result))
```
